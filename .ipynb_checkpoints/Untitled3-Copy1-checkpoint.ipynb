```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
```

```python
test1 = pd.read_csv('C:\\Users\\dell\\Desktop\\Project Competitions\\Anlayticsvidhya\\TEST_FINAL.csv',parse_dates = True)
train1 = pd.read_csv('C:\\Users\\dell\\Desktop\\Project Competitions\\Anlayticsvidhya\\TRAIN.csv')
```

```python
# dropping ID and converting date to DateTime

train = train1.drop(['ID', ], axis =1)
test = test1.drop(['ID'], axis =1 )

test['Date'] = pd.to_datetime(test['Date'])
train['Date'] = pd.to_datetime(train['Date'])#train1['Sales'] = np.where(train1['Sales'] == 0,1,train1['Sales'])
train1.Store_id.nunique(), test1.Store_id.nunique()
```

```python
train['Sales'] = np.where(train['Sales'] == 0,1,train['Sales'])
train.Store_id.nunique(), test1.Store_id.nunique()
```

```python
train2 = train.copy()
total = pd.concat([train,test], axis =0)
total
```

```python
total.describe(include = 'all')
```

```python
import category_encoders as ce
from sklearn.preprocessing import StandardScaler

total['Store_id'] = total['Store_id'].astype('str')
total['Date'] = total['Date'].astype('str')

encoder=ce.TargetEncoder() 

x = StandardScaler().fit_transform(total[['Sales']])
y = encoder.fit_transform(total['Store_id'],x)
total['Storenc'] = y
total.drop(['Store_id','#Order'],axis =1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder, QuantileTransformer


encoder = LabelEncoder()

total[['Date','Store_Type','Location_Type','Region_Code','Discount']] = total[['Date','Store_Type','Location_Type','Region_Code','Discount']].apply(encoder.fit_transform)
total.head()
```

```python
from sklearn.model_selection import train_test_split

x = total.iloc[:len(train),].drop('Sales', axis =1)
y = total.iloc[:len(train),]['Sales']

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.3, random_state= 22)
```

```python
import statsmodels.api as sm

X_train_Sm= sm.add_constant(x_train)
ls=sm.OLS(y_train,X_train_Sm).fit()
print(ls.summary())
```

```python
from sklearn.linear_model import ElasticNet, Lasso,  BayesianRidge, BayesianRidge, Ridge
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor, ExtraTreesRegressor,AdaBoostRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_squared_log_error,r2_score
import xgboost as xgb

def model_pred(model):
    mdl = model()
    mdl.fit(x_train,y_train)
    y_pred = mdl.predict(x_test)
    print('accuracy on trainset for {} is'.format(mdl), mdl.score(x_train,y_train))
    print('accuracy on testset for {} is'.format(mdl), mdl.score(x_test,y_test))
    print('rsquare is ',r2_score(y_pred,y_test))
    print('mean_absolute_erroris', mean_absolute_error(y_pred,y_test))
```

```python
model_pred(RandomForestRegressor)
```

```python
from xgboost import XGBRegressor

model_pred(XGBRegressor)
```

```python
model_pred(GradientBoostingRegressor)
```

```python
model_pred(AdaBoostRegressor)
```

```python
from catboost import CatBoostRegressor

catcols = ['Date','Store_Type','Location_Type','Region_Code','Discount']

model = CatBoostRegressor()
model.fit(x_train,y_train,cat_features=catcols, silent= True)
```

```python
y_pred = model.predict(x_test)
print('accuracy on trainset for {} is'.format(model), model.score(x_train,y_train))
print('accuracy on testset for {} is'.format(model), model.score(x_test,y_test))
print('rsquare is ',r2_score(y_pred,y_test))
print('mean_absolute_erroris', mean_absolute_error(y_pred,y_test))
```
